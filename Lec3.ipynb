```python
Advertising_Investment=[10,15,20,25,30,35,40,45,50]
Sales_Growth=[25,30,40,45,50,60,65,70,80]
total_Advertising_Investment=sum(Advertising_Investment) ## sum(xi)
total_Sales_Growth=sum(Sales_Growth)## sum(yi)
total_Advertising_Investment_power=sum([x**2 for x in Advertising_Investment])#### sum(xi^2)
total_Advertising_Investmen_multi_Sales_Growth=sum([x*y for x,y in zip(Advertising_Investment,Sales_Growth)])## sum(xi*yi)
n=len(Advertising_Investment)
```

```python
m=(n*total_Advertising_Investmen_multi_Sales_Growth-(total_Advertising_Investment*total_Sales_Growth))/(n*total_Advertising_Investment_power-total_Advertising_Investment**2)
m
```

```python
b=(total_Sales_Growth-(m*total_Advertising_Investment))/n
b
```

```python
print(f"y={m}*x+{b}")
x_n_1=60
y_n_1=100
print(f"Sales Growth for 60k ILs(y(n+1))={(x_n_1*m)+b}")
print(f"for 100K IL Sales Growth need to Invest(x(n+1)) ={(y_n_1-b)/m}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Our data
advertising_Investment_d = np.array(Advertising_Investment).reshape(-1, 1)  # Study hours
sales_Growth_d = np.array(Sales_Growth)  # Exam scores

# Create regression model
model = LinearRegression()
model.fit(advertising_Investment_d, sales_Growth_d)

# Print results
print(f"Slope (m): {model.coef_[0]:.2f}")
print(f"Intercept (b): {model.intercept_:.2f}")

# Calculate equation
equation = f"y = {model.coef_[0]:.2f}x + {model.intercept_:.2f}"
print(f"Line equation: {equation}")

# Predict hours needed to get score of 100
# score_to_predict = 100
# hours_needed = (score_to_predict - model.intercept_) / model.coef_[0]
# print(f"To get a score of 100, approximately {hours_needed:.2f} hours of study are needed")

# # Create the graph
plt.figure(figsize=(10, 6))
plt.scatter(advertising_Investment_d, sales_Growth_d, color='blue', label='Data points')
plt.plot(advertising_Investment_d, model.predict(advertising_Investment_d), color='red', label='Regression line')

# # Add prediction point
# plt.scatter([[hours_needed]], [100], color='green', s=100, label='Our prediction')

# Add labels in English
plt.title('Linear Regression - Advertising_Investment vs. Sales_Growth')
plt.xlabel('Advertising_Investment')
plt.ylabel('Sales_Growth')
plt.grid(True)
plt.legend()

# Display equation on the graph
plt.text(1, 95, equation, fontsize=12)

plt.show()
```
